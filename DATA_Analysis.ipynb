# Bad_droids

```python
# Q1 analiza = Q1 (Responses) - analizy
BD = pd.read_excel(io = "BAZA DANYCH - główna.xlsx")

# Q2 analiza = Q2 (Responses) - analizy
bad_d = pd.read_excel(io = "bad_droids.xlsx")
```

```python
# "Exp_noaudio"

x = 1
for idx1 in BD.index:
    for idx2 in bad_d.index:
        if BD.loc[idx1, 'mail1'][1:9] == bad_d.loc[idx2, 'username'][1:9] and BD.loc[idx1, 'GROUP'] == "Exp":
            BD.loc[idx1, 'GROUP'] = "Exp_noaudio"
            print("nr {}: ".format(x), idx1, " and ", idx2, BD.loc[idx1, 'mail1'], "---------", bad_d.loc[idx2, 'username'])
            x += 1
```

```python
# "Placebo_bad"

x = 1
for idx1 in BD.index:
    for idx2 in bad_d.index:
        if BD.loc[idx1, 'mail1'][1:9] == bad_d.loc[idx2, 'username'][1:9] and BD.loc[idx1, 'GROUP'] == "Placebo" and BD.loc[idx1, 'NIGHTS'] <= 5:
            print("nr {}: ".format(x), idx1, " and ", idx2, BD.loc[idx1, 'mail1'], "---------", bad_d.loc[idx2, 'username'])
            BD.loc[idx1, 'GROUP'] = "Placebo_bad"
            x += 1
```

```python
# "Eksp_bad"

x = 1
for idx1 in BD.index:
    for idx2 in bad_d.index:
        if BD.loc[idx1, 'mail1'][1:9] == bad_d.loc[idx2, 'username'][1:9] and BD.loc[idx1, 'GROUP'] == "Exp" and BD.loc[idx1, 'NIGHTS'] <= 5:
            print("nr {}: ".format(x), idx1, " and ", idx2, BD.loc[idx1, 'mail1'], "---------", bad_d.loc[idx2, 'username'])
            BD.loc[idx1, 'GROUP'] = "Exp_bad"
            x += 1
```

```python
# "Control"

x = 1
for idx1 in BD.index:
    if (BD.loc[idx1, 'GROUP'] == "Exp" or BD.loc[idx1, 'GROUP'] == "Placebo") and BD.loc[idx1, 'NIGHTS'] <= 5:
        print("nr {}: ".format(x), idx1, BD.loc[idx1, 'mail1'])
        BD.loc[idx1, 'GROUP'] = "Control"
        x += 1
```

```python
BD.to_excel("BAZA DANYCH.xlsx")
```
